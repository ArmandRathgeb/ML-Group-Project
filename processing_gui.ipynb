```python
import io
import cancerclass as cacl
import ipywidgets as widgets
import pandas as pd 
from IPython.display import display, clear_output
```

```python
# Load data 
output = widgets.Output()
labels = None
df     = None 

uploader = widgets.FileUpload(accept='.csv', multiple=True, description='Upload data and label files')

def handle_upload(change):
    global labels, df
    with output:
        clear_output()

    if not uploader.value:
        print("Please select a file")
    if len(uploader.value) < 2:
        print("Please select a data file and then the label file")

    data_file_if = uploader.value[0]
    label_file_if = uploader.value[1]

    try:
        data_content = io.StringIO(data_file_if.content.decode('utf-8'))
        label_content = io.StringIO(label_file_if.content.decode('utf-8'))
        df = pd.read_csv(data_content)
        labels = pd.read_csv(label_content)
        labels = cacl.ohe_data(labels)
    except Exception as e:
        print(f"Error reading data: {e}")

uploader.observe(handle_upload, names='value')
display(uploader,output)
```
